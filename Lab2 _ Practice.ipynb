### 2. Preprocess data
Preprocess the dataset as you have done before

##### 2.1 Binary encoding
Use .replace to replace the following columns:

- y
- default
- housing
- loan

from string to number

```python
bankData.head()
```

```python
replace_word = {'no': 0,'yes':1}
bankData_1 = bankData.replace(replace_word )
bankData_1.head()
```

#### 2.2 Convert categorical variables into dummy columns

(1) Use pd.get_dummies to convert the following categorical variales into dummy columns
- job
- maritial
- education
- contact
- month
- poutcome

```python
bankData_1.head()
```

```python
bankData_2 = pd.get_dummies(bankData_1, columns=['job','marital','education','contact','month','poutcome'])
bankData_2.info()
```

#### 2.3 Train/Test separation

Perform hold-out method
- separate feature / label
- 60% training set
- 40% testing set

use sklearn to separate data 
- more info : https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
#featur / label

label_bit = bankData_2.y.copy()
label_bl = bankData_2.balance.copy()
feature = bankData_2.drop(['y','balance'],axis=1).copy()
```

```python
from sklearn.model_selection import train_test_split

#for bit data
X_train_bit, X_test_bit, y_train_bit, y_test_bit = train_test_split(feature,label_bit,test_size=0.4,random_state=42)
```

```python
X_train_bl, X_test_bl, y_train_bl, y_test_bl = train_test_split(feature,label_bl,test_size=0.4,random_state=42)
```

### 3. Model Training 
We will use Scikit-learn package to build a model.
- http://scikit-learn.org/stable/

```python
from sklearn import linear_model,tree

model_logistic = linear_model.LogisticRegression()
model_tree = tree.DecisionTreeClassifier()
model_linear = linear_model.LinearRegression()
```

- Fit Logistic Regression model

```python
model_logistic.fit(X_train_bit, y_train_bit)
```

- Fit Decision Tree model

```python
model_tree.fit(X_train_bit, y_train_bit)
```

- Fit Linear Regression model

```python
model_linear.fit(X_train_bl, y_train_bl)
```

### Part 4: Model Evaluation

Evaluation metrics for classification
- confusion metrix
- accuracy
others :: - precision, recall, f1-score

Evaluation metrics for Linear Regression 
- R2

*** special :: visualization decision tree ***

#### Evaluation Model Logistic Regression 

```python
y_pre_logistic = model_logistic.predict(X_test_bit)

from sklearn.metrics import accuracy_score
print('Accuracy: %s'%accuracy_score(y_test_bit, y_pre_logistic ))

from sklearn.metrics import confusion_matrix
print('Confusion Matrix')
print(confusion_matrix(y_test_bit, y_pre_logistic  ))
```

#### Evaluation Model Decision Tree

```python
y_pre_tree = model_tree.predict(X_test_bit)

from sklearn.metrics import accuracy_score
print('Accuracy: %s'%accuracy_score(y_test_bit, y_pre_tree))

from sklearn.metrics import confusion_matrix
print('Confusion Matrix')
print(confusion_matrix(y_test_bit, y_pre_tree ))
```

#### Evaluation Model Linear Regression 

```python

y_pre_linear = model_linear.predict(X_test_bl)

from sklearn.metrics import r2_score
print('R2: %.2f'%r2_score(y_test_bl, y_pre_linear))
```

#### Tree Visualization

```python
import graphviz
dot_data = tree.export_graphviz(model_tree, out_file=None, 
                              feature_names=X_train_bit.columns,
                              class_names=['0','1'],
                              filled=True, rounded=True,
                              special_characters=True, rotate=True)
graph = graphviz.Source(dot_data)
graph.render('dtree_render')
```
